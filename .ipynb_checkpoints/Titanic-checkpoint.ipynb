```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

# Open Data CSV #####

```python
test_titanic = pd.read_csv('test.csv')
gender_sub = pd.read_csv('gender_submission.csv')
train_titanic = pd.read_csv('train.csv')
test_titanic
```

```python
gender_sub
```

```python
train_titanic
```

# Del Columns and Remove NAN lines from Training Data

```python
train_titanic = train_titanic.drop(['Name','Ticket','Fare','Cabin','Embarked'], axis=1)
train_titanic = train_titanic.dropna(how='any', axis=0)
train_titanic
```

# Check for NaN values in Testing Data

```python
test_titanic.isna().sum()
```

# Calculate the age´s mean to replace with NaN

```python
mean = test_titanic['Age'].mean()
```

# Remove unnecessary columns and fill Nan with Age mean on Testing Data 

```python
test_titanic = test_titanic.drop(['Name','Ticket', 'Fare','Cabin', 'Embarked'], axis=1)
test_titanic = test_titanic.fillna(mean)
test_titanic
```

# Change de column type

```python
train_titanic = train_titanic.astype('object')
train_titanic[['Age','PassengerId']]=train_titanic[['Age','PassengerId']].astype('int')
train_titanic.dtypes
```

```python
test_titanic = test_titanic.astype('object')
test_titanic[['PassengerId','Age']]=test_titanic[['PassengerId','Age']].astype('int')
test_titanic.dtypes
```

# converting categorical data into dummy or indicator variables (Training/Testing)

```python
train_titanic = pd.get_dummies(train_titanic,drop_first=True)
train_titanic
```

```python
test_titanic = pd.get_dummies(test_titanic,drop_first=True)
test_titanic
```

# check number os columns in each data and remove two extra columns from testing

```python
train_titanic.columns
```

```python
test_titanic.columns
```

```python
test_titanic = test_titanic.drop(['SibSp_8','Parch_9'],axis=1)
test_titanic
```

```python
ax = plt.subplots(figsize=(15, 13))
ax = sns.heatmap(train_titanic.corr(), annot=True, cmap="Wistia")
plt.show()
```

# Machine Learning - Separate columns X,Y and training by using DecisionTreeClassifier

```python
from sklearn.model_selection import train_test_split

y = train_titanic["Survived_1"]
x = train_titanic.drop("Survived_1", axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo_decisao = DecisionTreeClassifier()
modelo_decisao.fit(x,y)
```

# Criating predict with testing data

```python
previsao_classificacao = modelo_decisao.predict(test_titanic)
```

```python
print(previsao_classificacao)
```

# Subs the values from 0 to 1. The predicion generate results for sex Male and we are comparing with Female

```python
aux = len(previsao_classificacao)
for x in range(aux):
    if previsao_classificacao[x]==0:
        previsao_classificacao[x]=1
    elif previsao_classificacao[x]==1:
        previsao_classificacao[x]=0
```

```python
print(previsao_classificacao)
```

# Creating Dataframe to compar the prediction with kaggle results.

```python
tabela_comparar = pd.DataFrame()
tabela_comparar['previsao'] = previsao_classificacao
tabela_comparar['PassengerId_P']= test_titanic['PassengerId'].astype('int')
tabela_comparar['resultado'] = gender_sub['Survived']
tabela_comparar['PassengerId_R']=gender_sub['PassengerId']
tabela_comparar
```

```python
tabela_comparar['previsao'].value_counts()
```

```python
tabela_comparar['resultado'].value_counts()
```

```python
submission = pd.DataFrame(tabela_comparar,columns=['PassengerId_P','previsao'])
submission
```

```python
submission = submission.rename(columns={'PassengerId_P': 'PassengerId', 'previsao':'Survived'})
```

```python
submission.to_csv("submission.csv",index=False)
```
